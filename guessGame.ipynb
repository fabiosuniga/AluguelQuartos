```python
from random import shuffle

def guess_number():
  num = ''
  while num not in ['1', '2', '3']:
    num =  input('Guess the position between [1 - 2 - 3]: ')
  return int(num)

def shuffle_list(list_game):
  shuffle(list_game)
  return list_game

def guess_compared(guess, final_list):
  if final_list[guess-1] == 'O':
    print('Your guess is correct!')
    print(final_list)
  else:
    print('Your guess is incorrect!')
    print(final_list)

game_list = ['', 'O', '']

game_list = shuffle_list(game_list)
player_guess = guess_number()

guess_compared(player_guess, game_list)


```
